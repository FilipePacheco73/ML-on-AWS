# Task 2: Upload the data to S3 bucket

```python
import sagemaker


# Creating Sagemaker session using Sagemaker SDK
sess = sagemaker.Session()

#Prefix that we want to use

prefix = 'auto-pilot/used-car-data/input'

#Uploading the data to S3 bucket
uri = sess.upload_data(path = 'bike_sharing_daily_1.csv', key_prefix = prefix)

print(uri)
```

```python
# location to store the output

output = "s3://sagemaker-us-east-1-279288473542/used-car-data/output"

print(output)
```
